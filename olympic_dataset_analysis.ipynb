## **Top 10 Country who having gold**

```python
asd = medals[medals['medal_type'] == 'GOLD']
top_10_gold_con = asd.groupby('country_name').count()['medal_type'].sort_values(ascending=False)[:10]

# Set the figure size
plt.figure(figsize=(8, 6))

# Create a horizontal bar plot with different colors
sns.barplot(x=top_10_gold_con.values, y=top_10_gold_con.index, palette="Set3")

# Add data labels to the bars
for i, value in enumerate(top_10_gold_con.values):
    plt.text(value, i, str(value), ha='left', va='center')

# Add axis labels and a title
plt.xlabel('Number of Gold Medals')
plt.ylabel('Country')
plt.title('Top 10 Countries with the Most Gold Medals')

# Display the plot
plt.show()
```

```python
gender_vs_model=medals.groupby(['event_gender','medal_type'])['country_name'].count().unstack()
gender_vs_model
```

```python
# Plot for finding the costly neighbourhood_group
ax = gender_vs_model.plot(kind='bar')

# Add labels to the bars
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    ax.annotate(f'{height}', (x + width/2, y + height), ha='center', va='bottom')

plt.xticks(rotation='horizontal')
plt.xlabel('Event Gender')
plt.ylabel('Count')
plt.title('Gender vs Medal Type')

plt.show()
```

```python
event_gender_counts = medals.groupby(['slug_game', 'event_gender'])['event_gender'].count().reset_index(name='count')

# Create a Seaborn bar plot to show event gender distribution across different slug games
plt.figure(figsize=(12, 6))
sns.barplot(x='slug_game', y='count', hue='event_gender', data=event_gender_counts)
plt.title('Event Gender Distribution by Slug Game')
plt.xlabel('Slug Game')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Event Gender')
plt.tight_layout()
plt.show()

# Filter and plot only the top 10 events based on the count of event genders
top_10_events = event_gender_counts.groupby('slug_game').apply(lambda x: x.nlargest(10, 'count')).reset_index(drop=True)

plt.figure(figsize=(12, 6))
sns.barplot(x='slug_game', y='count', hue='event_gender', data=top_10_events)
plt.title('Top 10 Event Gender Distribution by Slug Game')
plt.xlabel('Slug Game')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Event Gender')
plt.tight_layout()
plt.show()
```
